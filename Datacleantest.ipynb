```python
#handling the missing values
```

```python
for i in ["Hepatitis B","Polio","Adult Mortality"]:
    df[i].fillna(df[i].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for col in df.select_dtypes(include='number').columns:
    df[col]=impute.fit_transform(df[[col]])
```

```python
df.isnull().sum() #now no null values
```

```python
#outlier treatment
```

```python
def wisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in ['GDP','Schooling','Total expenditure']:
    lw,uw=wisker(df[i])
    df[i]=np.where(df[i]<lw,lw,df[i])
    df[i]=np.where(df[i]>uw,uw,df[i])
```

```python
#now no outlliers
for i in ['GDP','Schooling','Total expenditure']:
    sns.boxplot(df[i])
    plt.show()
```
